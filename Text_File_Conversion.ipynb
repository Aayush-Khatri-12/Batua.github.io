Plot the performance (time taken) of Python program for :

---> Convert 200MB, 400 MB, 600 MB, 800 MB, and 1000MB text files to upper case.

```python
import random as r
import string as s
base_limit_size=200*1024
number_of_files=5
for i in range(number_of_files):
  limit_size=(i+1)*base_limit_size
  name=f"randomQ4_{i+1}.txt"
  with open(name,'w') as f:
    cur_size=0
    while cur_size<limit_size:
      random_str="".join(r.choices(s.ascii_letters,k=15)) + "\n"
      str_size=len(random_str.encode('utf-8'))
      if (cur_size+str_size)>limit_size:
        break
      f.write(random_str)
      cur_size+=str_size
print("Done!!!")
```

```python
import time as t
import matplotlib.pyplot as plt
x=[]
y=[]
for i in range(number_of_files):
  name=f"randomQ4_{i+1}.txt"
  start=t.perf_counter()
  with open(name) as f:
    content=f.read()
  with open(name,'w') as f:
    f.write(content.upper())
  end=t.perf_counter()
  x.append(200*(i+1))
  y.append(end-start)
plt.plot(x,y)
plt.xlabel('size of file (in MB)')
plt.ylabel('Time taken to convert to upper case(in seconds)')
```
